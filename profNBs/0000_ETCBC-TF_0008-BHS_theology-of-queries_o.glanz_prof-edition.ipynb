## From Question...: Attributes of God
Lets say we want to find out how the divine names for YHWH (JHWH, >L, >LHJM) are being qualified. This would answer the question about how the character or person of YHWH is presented in the OT. Lets have a look at different examples that would help us formulate the actual linguistic question:

But there are other relations possible as well: regens-rectum, attributival, adjunctival, etc.


### `mother-daughter` relations: clause => clause
Lets now look at a third example:

![](https://5cqa6g.dm.files.1drv.com/y4mylMlMOFW83PBWG-msqf_LpBESVSOiTzlErg3sYWZEkmDd5c0iSVEHIeztZna3q2gGn1yA1BdvDHVZWZm39HHuXI6EY3JaNQG1ozx8KVARbVFYmZCDAjrCMn84ylMz2ZS_AMwX2YpleeGI5VxZ2qoPdz9EjbBG_PSQ-5xv2AnM9aCoIEmT2DGgQKxMY8AuZnwNjdy_dVtfNnHuljJ8C8jyQ/Annotation%202019-06-28%20112827.png?psid=1)


The above sample comes from Ex 6:7 and can be pulled up with this query:

```
ex6_7 = '''
verse book=Exodus chapter=6 verse=7
    sentence number=21
'''
ex6_7  = A.search(ex6_7)
A.show(ex6_7, start=1, end=2, condensed=True, colorMap={2:'magenta'})
```


From a linguistic perspective the above sentence consists of three clauses. We are particularly interested in the relation between the second and the third clause. The third clause is an attributive clause that specifies the entire previous clause functioning as the object of the first clause (cl1: "You know" => cl2: "That I am YHWH your God" => cl3: "who..."). Since the second clause is all about YHWH the third clause give us more information about YHWH. This type of clause relation is called a `mother-daughter` - clause relation. 

The daughter clause would be:
הַמֹּוצִ֣יא אֶתְכֶ֔ם מִתַּ֖חַת סִבְלֹ֥ות מִצְרָֽיִם

And it relates to its mother clause: כִּ֣י אֲנִ֤י יְהוָה֙ אֱלֹ֣הֵיכֶ֔ם

in an attributive way.

### `mother-daughter` relations: clause => phrase

Lets get to our fourth and final example coming from Gen 12:7
![](https://10fqka.dm.files.1drv.com/y4mo__Rp52UpyB1R5hYLwl-tHmiUv9scGsVFrkb9uQyLg3KibJeE4LReoL8c799oQjP56uopavAgMQfrBUSTlI5SCGVU6d-dxcy8woaYk1Zp7fGO2IWfdO5i7TuwhvqHqJIKkjNBK7edTJYYyOy6ebBXnHeZvLyn8x_AyxUFYRKwVeaVgcaz-ukIPjEQOFye72eb6VJkqcDyckU5hvXNeoMyg/Annotation%202019-06-28%20115215.png?psid=1)

Here we have a relative clause that refers back to a single phrase in the previous clause. In this case then, the `mother` of the `daughter` (relative clause: הַנִּרְאֶ֥ה אֵלָֽיו | "who appeared to him") is a phrase in the previous clause (לַיהוָ֖ה | "for YHWH" ).

Thus, clause-daughters can have phrase-mothers!


### Conclusion
The characterisation of a participant can be linguistically materialized in different ways. Having outlined the above four ways we are now going to execute queries that search for each of the above materializations. 

## ... To Query: Attributes of God

### ... by means of PreC in NCs

This query finds all nominal clauses (NC) in which YHWH/Elohim/El are attributed by means of an adjectival predicate compliment (the element that functions as a predicate in nominal clauses: God{Subj} (is) good {PreC}).

The MQL version of the query looks like this (https://shebanq.ancient-data.org/hebrew/query?version=2017&id=1125):


```
select all objects where
[clause kind = NC
    [UnorderedGroup
        [phrase FOCUS function = Subj
            NOTEXIST [word sp = prep  OR lex = "DBR/" OR lex = "XMH/" OR lex="JWM/" OR lex = "BJT/" OR lex = "JR>H/" OR lex="M>RH/" OR lex = "M<FH/"  OR lex = "QRBH/" OR lex = "PNH/"  OR lex = "<JN/"  OR lex = ">MRH/" OR lex = "DRK/"]
            [word lex = 'JHWH/' or lex = '>LHJM/' or lex = '>L/']
        ]
        [phrase FOCUS function = PreC
            [word sp=adjv]
        ]
    ]
]
```


````python
GodsPreCs1 = '''
clause kind=NC
    p1:phrase function=Subj
        w1:word lex=JHWH/|>LHJM/|>L/ 
        w2:word lex#DBR/|XMH/|JWM/|BJT/|JR>H/|M>RH/|M<FH/|QRBH/|PNH/|<JN/|>MRH/|DRK/ 
        w3:word sp#prep
    p2:phrase function=PreC
        w4:word sp=adjv lex_utf8*  lex*

w1 :: w2
w1 :: w3
p1 <  p2

'''
GodsPreCs1  = A.search(GodsPreCs1)
A.table(GodsPreCs1, start=1, end=2, condensed=False, colorMap={3: 'magenta', 7: 'yellow'})
A.show(GodsPreCs1, start=1, end=2, condensed=False, colorMap={3: 'magenta', 7: 'yellow'})
````

### ... by means of adjetival SubPhrases

This query finds all cases in which YHWH, Elohim, or El are attributed by adjectival sub-phrases (e.g. "And he was the priest of [God{mother-subphrase}, the most high{daughter-subphrase}]).

The MQL version of the query looks like this (https://shebanq.ancient-data.org/hebrew/query?version=2017&id=1124):

```
select all objects where
[phrase
	[subphrase FOCUS
		[subphrase as motherSubphrase
			[word lex = 'JHWH/' or lex = '>LHJM/' or lex = '>L/']
		]
	]
	..
	[subphrase FOCUS rela = atr AND mother HAS motherSubphrase.self]
]
```

````python
GodsAdjSub1 = '''
phrase
    sp1M:subphrase
        word lex=JHWH/|>LHJM/|>L/
    sp2D:subphrase  rela=rec|ajd|atr
        word sp=adjv lex_utf8* lex*

sp1M <mother- sp2D
'''
GodsAdjSub1  = A.search(GodsAdjSub1)
A.table(GodsAdjSub1, start=1, end=10, colorMap={3: 'magenta', 5: 'yellow'}, condensed=False)
# A.show(GodsAdjSub1, start=1, end=10, colorMap={3: 'magenta', 5: 'yellow'}, condensed=False)
````

### ... by means of relative clauses

This query finds all attributive clauses that refer to YHWH, Elohim, or El. Thus these attributive clauses are describing God in some further detail.

The MQL version of the query looks like this (https://shebanq.ancient-data.org/hebrew/query?version=2017&id=1121):

```
select all objects where
[sentence
	[
	[clause FOCUS   
		[clause as motherClause
			[word lex = 'JHWH/' or lex = '>LHJM/' or lex = '>L/']
		]
	]
	..
	[clause FOCUS rela = Attr AND mother = motherClause.self]
	]
	OR
	[
	[clause FOCUS   
		[clause as motherClauseG
			[word lex = 'JHWH/' or lex = '>LHJM/' or lex = '>L/']
		]
	]
	..
	[gap 
		[clause FOCUS rela = Attr AND mother = motherClauseG.self]
	]
	]
]
OR
[sentence
	[
	[clause
		[phrase FOCUS
			[phrase as motherPhrase
				[word lex = 'JHWH/' or lex = '>LHJM/' or lex = '>L/']
			]
		]
	]
	..
	[clause FOCUS rela = Attr AND mother = motherPhrase.self]
	]
	OR
	[
	[clause
		[phrase FOCUS
			[phrase as motherPhraseG
				[word lex = 'JHWH/' or lex = '>LHJM/' or lex = '>L/']
			]
		]
	]
	..
	[gap
		[clause FOCUS rela = Attr AND mother = motherPhraseG.self]
	]
	]
]
OR
[sentence
	[
	[clause 
		[word FOCUS
			[word as motherWord
				[word lex = 'JHWH/' or lex = '>LHJM/' or lex = '>L/']
			]

			]
		]
	]
	..
	[gap
		[clause FOCUS rela = Attr AND mother = motherWordG.self]
	]
	]
]
```

#### mother is a WORD with a divine name
