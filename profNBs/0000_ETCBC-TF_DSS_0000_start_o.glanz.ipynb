# Getting started with the TF-DSS corpus
This is the first of a series of jupyter notebooks that seek to introduce the TextFabric (TF) version of the Dead Sea Scrolls. The Dead Sea Scrolls have been studied for 70+ years now and made available successively through print volumes. All of the latest 32 volumes (of a total of 40 volumes) of the [Discoveries in the Judaean Desert](https://en.wikipedia.org/wiki/Discoveries_in_the_Judaean_Desert) were published with the Dutch-Isareli world-renowned textual critic [Emmanuel Tov](https://en.wikipedia.org/wiki/Emanuel_Tov) as editor. Since 1991 [Martin Abegg](https://en.wikipedia.org/wiki/Martin_Abegg) has worked with his partners on digitizing the DSS text and making them accessible to the wider scholarship in cooperation with Accordance Bible Software. Abegg's work became a milestone in DSS research as he broke open the 30+ years of research monopoly where only a selective circle of researchers were granted access to the facsimile of DSS texts. Wikipedia puts it this way: 
>Those who had the privilege of working with the scrolls were angry that their work had been undermined. However, Abegg's publication broke a forty-year monopoly on the scrolls, allowing scholars of all kind access to examine the artifacts. Furthermore, Abegg's work on the concordance gave him the opportunity to create computer research tools in association with Accordance Bible Software for Dead Sea Scrolls research. Essentially, Abegg's work accelerated Dead Sea Scrolls research. (https://en.wikipedia.org/wiki/Martin_Abegg)

The TF-DSS app is part of the CACCHT (Creating Annotated Corpora of Classical Hebrew Texts) project funded by the Charis Foundation. The project is chaired by Dr. Wido van Peursen [Free University of Amsterdam](https://research.vu.nl/en/persons/wido-van-peursen), Dr. Robert Rezetko [University of Syndey, Radboud University Nijmegen](https://independent.academia.edu/RobertRezetko), and Dr. Oliver Glanz [Andrews University](https://www.andrews.edu/sem/faculty_staff/faculty/oliver-glanz.html). The analytic work is carried out by Jarod Jacobs (Warner Pacific College) and Martijn Naaijer [Eep Talstra Center for Bible and Computer](http://etcbc.nl/). Most of the programming involved with the TF implementation is done by Dr. Dirk Roorda (Royal Dutch Academy of the Sciences, https://dans.knaw.nl/). The CACCHT team uses Abegg's (archived at the Software Heritage Archive). The original Abegg data is then restructured and updated so that it can be made available via TF. As a next step, the data is linguistically analyzed with the help of the ETCBC research infrastructure. The team seeks to provide a text-driven, bottom-up analysis of all linguistic levels (morphology, phraseology, syntax, text-grammar). Due to the fragmentary nature of many DSS texts, this can only partially implemented.
The CACCHT TF-DSS work is made available under the MIT license at https://github.com/ETCBC/dss. For a description about the project got [here](https://github.com/ETCBC/dss/blob/master/docs/about.md).

Lets look at the first line of that scroll. In order to do so, we define a query that we call "FirstLineOf1Qisaa" (any random name would do). Then we write the actual query:

1. we define the scroll as `scroll scroll=1Qisaa` 
2. we want to have a look the first column with `fragment fragment=1`
3. we define the the first line as `line line=1`

And then execute the query with `FirstLineOf1Qisaa=A.search(FirstLineOf1Qisaa)`

````python
FirstLineOf1Qisaa ='''
scroll scroll=1Qisaa 
 fragment fragment=1
  line line=1
'''
FirstLineOf1Qisaa=A.search(FirstLineOf1Qisaa)
````

## Searching within a particular Qumran scroll/fragment: the great Isiah scroll *1Qisaa*
Lets start again with the Great Isiah scroll. Before we can search we need to figure out what features we can look for. An overview over all features can be retrieved with the command `TF.featuresSets`.

````python
TF.featureSets
````

A comprehensive description of all features can be found here: https://github.com/ETCBC/dss/blob/master/docs/transcription.md#morphological-features. 

We will start with the lexical features `lexe` and `lexo`. 
- **lexo** represents the original Abegg's transcription ("o" stands for "original").
- **lexe** reprensets the ETCBC version of the transcription ("e" stands for "ETCBC").

Let now look for the lexemes (`word lex*`) of the first line (`line line=1`) of column I (`fragment fragment=1`) of the Great Isiah Scroll (`scroll scroll=1Qisaa`):

````python
Q ='''
scroll scroll=1Qisaa
    fragment fragment=1
        line line=1
            word lexe lexo
'''
````

````python
R=A.search(Q)
````

````python
A.show(R, start=1, end=10, withNodes=True, colorMap={1: 'yellow', 2: 'cyan', 3: 'magenta', 4: 'green'},fmt='layout-orig-full')
````

# Looking for Biblical materials
For textual criticism and general historical research it is often important to find out whether bibical Qumran fragments deviate from the MT. How can you find all Qumran texts that contain the bible verse you are working on?

In the following query one finds all scrolls that contain Gen 1:1.
We are looking for any scroll `scroll` that contains a word that comes from the biblical text of Gen 1:1 `word book=Gen chapter=1 verse=1`.

````python
TheOTinDSS = '''
scroll
    word book=Gen chapter=1 verse=1
'''
TheOTinDSS  = A.search(TheOTinDSS)
A.table(TheOTinDSS, start=1, end=100, condensed=True, fmt='layout-orig-full')
````

There are 27 words of coming from Gen 1:1 found in three different Qumran texts!
And Jeremiah 4:5? Does that verse appear in DSS texts?

````python
Bible2 = '''
line
  word book=Jer chapter=4 verse=5
'''
Bible2  = A.search(Bible2)
A.table(Bible2, start=1, end=4, condensed=True, fmt='layout-orig-full')
````

Yes, in two fragments!

# Searching in language specific DSS materials (Hebrew, Aramaic, Greek)
The DSS material contains scrolls and fragments in three different languages: Hebrew, Aramaic, and Greek. These languages are values of the features `lang` of the objecttype `word`. The values available are `a` for Aramaic and `g` for Greek. The default value is Hebrew and does not need to be specified. Looking for Greek words in scrolls/fragments could be done in the following way:

# Searching Biblical and Non-Biblical DSS material
The TF-DSS app allows you to search in fragments and scrolls that contain only biblical, only non-biblical, or a mixture of both materials. THe different categories are values of the feature `biblical` as one casen see below:  (https://github.com/etcbc/dss/blob/master/docs//transcription.md#biblical-or-not-biblical):


## Scrolls/fragments with biblical materials only
There are many many fragments that contain biblical materials. Let's look for phrases that contain something like "dying (MWT) soul (NPC)". We want both words to appear in the same line and follow each other without another word in between `w1 <: w2`. We want to look only in scrolls/fragments that contain biblical material `scroll biblical=1|2|3|4`.

````python
BiblicalColumns = '''
scroll biblical=1|2|3|4
    line
        w1:word lexe~M.W.T.*
        w2:word lexe~N.P.C.*
w1 <: w2
'''
BiblicalColumns  = A.search(BiblicalColumns)
A.table(BiblicalColumns, start=1, end=25, condensed=True, fmt='layout-orig-full')
````

There are three Qumran fragments that contains biblical material and have the phrase "dying soul".

## Scrolls/fragments with no biblical materials present `biblical#1|2|3|4` 
Lets look for the word NPC in scrolls/fragments that does not contain any biblical material.

````python
NoBiblicalMaterial = '''
scroll biblical#1|2|3|4
    line
        word lexe~N.P.C.*
'''
NoBiblicalMaterial  = A.search(NoBiblicalMaterial)
A.table(NoBiblicalMaterial, start=1, end=2, condensed=True, fmt='layout-orig-full')
````

The Damascus Document contains the word NPC.

## Biblical Material in scrolls/fragments that holds both Biblical and non-Biblical materials
In the following query we are going to look for words that are from the biblical canon `word biblical=1` but appear in DSS scrolls/fragments that contain both biblical and non-biblical material `scroll biblical=2`.

````python
BiblicalInMixedMaterial = '''
scroll biblical=2
    line
        word biblical=1
'''
BiblicalInMixedMaterial  = A.search(BiblicalInMixedMaterial)
A.table(BiblicalInMixedMaterial, start=1, end=2, condensed=True, fmt='layout-orig-full')
````
