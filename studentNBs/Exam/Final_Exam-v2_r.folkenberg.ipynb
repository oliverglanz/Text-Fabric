### 3. Determing the Translation of the First Clause of [1 Kg. 17:22 (BHS)](https://accordance.bible/link/read/ESVS#1Kings_17:22)
In the below query I have searched specifically for the 4 most dominant valence constructions, and will use the findings to determined how the first clause of [1 Kg. 17:22 (BHS)](https://accordance.bible/link/read/ESVS#1Kings_17:22) should be translated.

````python
HearVal2 = '''
clause
    phrase function=Pred
        word lex=CM<[
    phrase function=Cmpl|Objc
        =: word lex=>T|>L|B|L
'''
HearVal2  = A.search(HearVal2)
A.table(HearVal2, start=1, end=20, condensed=False, extraFeatures='qere_utf8', fmt='text-orig-full-ketiv', colorMap={3:'yellow', 5:'magenta'})
````

From an analysis of the above results, the valence most accurate translation can be determined.

When used with >T, to "hear" is referring to literally hearing something and not to any form of response.

When used with >L, the "hear" is reffering to not only literally hearing, but to responding in the affirmative/obeying what the individual or people have heard.

When used with B, the reference is similar to that of L. The point of focus is not the literaing hearing but of the obeying what has been heard.

When occuring with L, to "hear" carries the meaning of listening and obeying, similar to >L.

#### Conclusion:
After anaylzing the valence of "hear" in the first clause of 1Kg. 17:22, the presence of "B" in this clause means the translation should be as follows:
"And the LORD obeyed the voice of Elijah."

## Perfect/Qatal "T" endings as non 2sgF
Sometimes, translations, scholars, databases and commentaries deviate from the paradigmatic interpretation of verbal forms. Fortunately, the ETCBC database allows scholars to search for both the graphical forms of verbal forms with the word feature `g_cons` and the morphological interpretations by the features `ps`,  `nu`,  `gn`, `vt` etc...


In combination with regular expressions (initiated by `~`) and `g_cons` we can search for exactly the verbal forms we are interested in. For example, the typical T ending of a 2sgF perfect/qatal form could be found with

```
word g_cons~.*T$ ps#p2 gn#F vt=perf qere* g_word_utf8* language=Hebrew
```

In this case we are looking for the ending "T" ("T" stands at the end as it is followed by the regular expression `$` marking the end of the word) in ver perfect/qatal verbal form that is NOT morphologically interpreted as 2nd person (`ps#2`) feminine (`gn#F`).

If we distinguish between ketiv, qere and orthographic variations its handy to compare the different display options of `fmt`:
