```python
# First, I have to laod different modules that I use for analyzing the data and for plotting:
import sys, os, collections
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt; plt.rcdefaults()
from matplotlib.pyplot import figure
from collections import Counter

# Second, I have to load the Text Fabric app
from tf.fabric import Fabric
from tf.app import use
```

```python
A = use('bhsa', hoist=globals())
```

# QHL
## QHL as subject

```python
QHLsubject='''
clause
 phrase function=Subj
   word lex_cons~.*[Q].*[H].*[L].* sp=verb|subs lex* lex_utf8*
 phrase function=Pred|PreO|PreC|PreS
   word lex* lex_utf8* sp=verb

'''
QHLsubject  = A.search(QHLsubject)
A.table(QHLsubject, start=1, end=10, condensed=False, extraFeatures=("g_word_utf8"), colorMap={2: 'magenta'})
```

```python
A.export(QHLsubject, toDir='D:/OneDrive/1200_AUS-research/Fabric-TEXT/PhD_rodrigo.galiza', toFile='QHLsubject.tsv')
```

```python
QHLsubject=pd.read_csv('D:/OneDrive/1200_AUS-research/Fabric-TEXT/PhD_rodrigo.galiza/QHLsubject.tsv',delimiter='\t',encoding='utf-16')
pd.set_option('display.max_columns', 50)
QHLsubject.head(11)
```

```python
QHLsubject.groupby(["lex_utf85"]).count()
```

```python
sns.lmplot(x="S1", y="R", data=QHLsubject, hue='lex_utf85', height=5, aspect=7/3, fit_reg=False, scatter_kws={"s": 200})
ax = plt.gca()
ax.set_ylabel('distribution of verbal predicates')
ax.set_xlabel('OT books')
```
