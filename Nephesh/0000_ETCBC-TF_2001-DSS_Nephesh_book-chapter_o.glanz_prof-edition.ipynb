## Findings in Dead Sea Scrolls Study edition (non biblical)
After checking all instances of NPC in the non-biblical Qumran scroll we cannot find a single instance that assumes a framework that knows of a seperation of soul and body. The soul is never removed from the body. Rather, the soul is removed from a sinful living to a holy living.
Salvation of the soul is never salvation from the body but salvation from concrete enemies (wicked people, danger) or immorality.
- "my soul will not crave (תאוה!) wealth by violence" (https://ref.ly/logosres/brdssseen?ref=DSSSE.1Q27+Frag.+1+i%3a4&off=38181)
