## The Dictionaries...
It is a typical feature of languges that many of their nouns are derived from verbs. E.g. the noun "search" in "the search for gold was painful" is derived from the verb "to search". This is no different to Hebrew where many nouns, including NPC/. In the TNK the verb NPC[ appears only in its niphal form with the meaning "to breath freely"  (cf. HALOT (https://ref.ly/logosres/hal?ref=Page.p+711&off=1935). We can, however, assume that the standard qal stem meaning must be "to breath". Thus, the meaning is not just to take in air and release it again in order to sustain life. It rather refers to a particular type of breathing that could be best rendered in english as "to breath freely" or "to breath without restrictions". This is clear through the insights of comparative linguistics. The same root NPC appears in Arabic (nafasa) and Akkadian (napāšu) also with the meaning "to breath (freely)".

Usually human beings are not aware of their breathing activity, it comes naturally and automatic. It is only at the moment where a sickness has caused a swelling of the throat that we realize that there is such a thing as breathing *freely*. It does not suprise that the nominal NPC form refers in Akkadian primarily to the throut (napištu). In Amorite it means "breath" (napš ). In both languages the concrete meaning of NPC is then used to refer to the more abstract concept of "life" (https://ref.ly/logosres/hal?ref=Page.p+711&off=3444&ctx=throat%2c+life%3b+Amor.+~naps%CC%8C+breath%2c+life+). The rational is quite obvious: when breathing comes to an end, life comes to an end. THus the most primary need and desire of existnce is breathing. NPC is therefore considered the most important organ and a symbol for vitality (see A. R. Johnson Vitality 3ff; Sander ZAW 77 (1965):329ff; D. Lys Nephesh (Paris 1959); Reicke-R. Hw. 1755f; Wolff Anthropologie 25ff; THAT 2:71ff and 72). While once can fast for some time, abstain from sex or sleep for a while, One cannot stop breathing for more than a few minutes. Thus, breathing becomes the most central desire of human existance. This desire is so fundamental that the word NPC is also used as a synonym for desire.


NPC/ itself, then becomes a symbol for the unity and integrity of the person.

In all cases the Nepesh is being destroyed or cut off. The formulation is used as an idiomatic expression for "being killed". This  becomes clear through the fact that generaly humans can be cut off. We find both the formulation >JC/ being cut off (cf. 1 Kgs 2:4) as well as NPC/ being cut off.

#### XJH[
After KRT[ the verbs XJH[, HJH[, and >WH[ appear each 8x. While the Nephesh can be cut off, i.e. a human being dies, the Nephesh can also be alive. Most of the time XJH[ appears as a jussive renering the meaning "the soul shall live" (e.g. Gen 19:20, 1 Kgs 20:32, Isa 55:3, Jer 38:20, Psa 119:175). Thus, the wish for a soul to live assumes that the soul is vulnerable and threatened by death.

#### HJH[
The impression of Nepesh possibly dying or surviving is also present in its valence construction with HJH[. 1 Sam 25:29, describes how one can seek after (BKC[) one soul or sling away (QL<[) a soul as an idiomatic expression for "killing". As contrast to BKC[ and QL<[ the soul can "be"(HJH[), i.e. remain its existance. 
The two appearances of the construction in 1 Kings appear in the same context: The nephesh of one person can be used as the revenage for the lost (dead) nephesh of another person. 
3 of the 4 construction with HJH[ in Jeremiah continue the impression: Here the Nephesh is at risk to perish but the Lord promises the possibility of naked survival considering the fall of Jerusalem. While everything will get lost there is a chance that the Nephesh could survive (Jer 21:9, 38:2, 39:18).

#### >WH[
In contrast to the other three verbs, the valance with >WH[ expresses genrally a wish for well tasting food. At the core it describes taste, as a primary food related sensation. However, derived from this one can also have a taste for power or control.... That the primary meaning of the construction has to do with food becomes clear when one look at all constructions with >WH[. In most of the cases, food is the object of longing/craving for: Num 11:4, 34; 2 Sam 23:15; Prov 23:3, 6: 1 Chr 11:17. From this basic meaning other uses of >WH[ are developed. For example to long of riches, etc.

WHen one does not only look at syntactical subject-predicate relations one realizes that the default construction of >WH[ comes with NPC/. Thus, there, where NPC/ is missing it is replaced by a person (e.g. David). Form this, one can conclude that NPC/ can stand for a person in tota.

#### words of gusto
In addition to the frequently used >WH[ constructions a comparison with other verbs that have NPC/ as subject creates a fuller image of the Sachverhalt. 
1. a>KL[: The NPC can eat.
2. FB<[: The NPC/ can be sated (gesaettigt). The verb has as primary function the idea of food saturation. E.g. Ex 16:8, 12; jer 44:17. That is the reason why it is often preceded by the word >KL[ (e.g. Lev 26:26, Deut 6:11, 8:10, 12, 26:12, Isa 44:16, Hos 4:10). From this primary food related meaning other meanings like being staurated with peace and meaning (see here particularly Isa 58:11, 66:11 who makes the link with food explicit, but others do not establsihe that link: Jer 31:14, ).
3. R<B[ if the NPC/ can eat and can be satisfied with food it becomes clear that it an also starve when it does not receive enough food. Although its use in Prov 19:15 is metaphorical as NPC/ is not starving from food but from meaning the primary usage of R<P[ is food related (Gen 41:55, Isa 9:19, Jer 42:14). Dan 8:3 makes this connection explicit.
4. YM>[ is the partner word of R<P[ as it expresses thirst in contrast to hunger. Obviously NPC/ can drink and eat when it can be thirsty and hungry. Both appearance in the Psalms (Psa 42:3, 63:2), however, do not have fluids in mind but God as the object that can still the thirsty NPC/. Thus, we can see again, that the Nephesh appears both as a biological organ that can eat and drink but is also used as an entity that metaphorically eats and drinks God and wisdom.
4. XPY[, RYH[ These words express satisfaction, fullfillment, and saturation. This does not come as a surprise considering the previous findings.

3. "Frequently in the Old Testament nepesû designates the individual (Lev. 17:10; 23:30). More significant here is that nepesû can mean the corpse of an individual even without the qualification “dead” (Num. 5:2; 6:11). Here nepesû is detached from the concept of life and refers to the corpse. Hebrew thought could not conceive of a disembodied nepesû. (Carl Schultz, “Soul,” in Evangelical Dictionary of Biblical Theology, electronic ed., Baker reference library (Grand Rapids: Baker Book House, 1996), 743.)
5. "Frequently nepesû takes the place of a personal or reflexive pronoun (Ps. 54:4; Prov. 18:7). Admittedly this movement from the nominal to the pronominal is without an exact borderline. The Revised Standard Version reflects the above understanding of nepesû by replacing the King James Version “soul” with such translations as “being,” “one,” “self,” “I/me.” (Carl Schultz, “Soul,” in Evangelical Dictionary of Biblical Theology, electronic ed., Baker reference library (Grand Rapids: Baker Book House, 1996), 743.)
6. "It can refer to the throat (Isa. 5:14; Hab. 2:5), noting that it can be parched and dry (Num. 11:6; Jer. 31:12, 25), discerning (Prov. 16:23), hungry (Num. 21:5), and breathing (Jer. 2:24). Nepesû also can mean the neck, and the vital function that takes place there, noting that it can be ensnared (1 Sam. 28:9; Ps. 105:18), humbled and endangered (Prov. 18:7), and bowed to the ground (Ps. 44:25). Even while focusing on a single part of the body, by synecdoche the whole person is represented. (Carl Schultz, “Soul,” in Evangelical Dictionary of Biblical Theology, electronic ed., Baker reference library (Grand Rapids: Baker Book House, 1996), 743.)
7. "Instead of splitting a person into two or three parts, Hebrew thought sees a unified being, but one that is profoundly complex, a psychophysical being. (Carl Schultz, “Soul,” in Evangelical Dictionary of Biblical Theology, electronic ed., Baker reference library (Grand Rapids: Baker Book House, 1996), 743–744.)
