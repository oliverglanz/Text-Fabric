# A brief history of research tools
from: **Glanz, Oliver. “Bible Software on the Workbench of the Biblical Scholar: Assessment and Perspective.” Andrews University Seminary Studies 56, no. 1 (2018): 5–46.**

>


Database producers of biblical Hebrew and Greek often approach their texts and digital tools in a different way than most users of Bible Software. While database producers search for linguistic patterns from the smallest units (phonology: sound units) up to the highest language structures (text-grammar: grammatical backbone of texts) , Bible Software users predominantly use their databases as a digital extension of their analog tools (Hebrew/Greek texts, dictionaries, concordances, grammars). [...]

Some 23 years ago, in 1991, the first popular commercial version of Logos Bible Software was published, followed by BibleWorks in 1992 and Accordance in 1994.  The United Bible Society published their first non-commercial Bible-translation Software in 1997.  Digitalizing the biblical print medium had several advantages. The most obvious benefits are the following:

1.	Saving of physical space => mobility of libraries
2.	Speeding up reading => linking of biblical words with dictionary entries. Lowering the bar for dealing with original languages: for example, with the linking of data sets (grammar, dictionary, source text) the user no longer had to connect lexical information with concrete morphological realizations. Bible Software users who are not able to read Hebrew and Greek are still able to look up the meaning of words in dictionaries without knowing different alphabets, spellings, and grammar (declinations, conjugations).
3.	Searching for words and word combinations => flexible electronic concordance. Classical, frequently used, and well trusted concordances  became unnecessary since Bible Software was able to find the distribution of each single word in an instant.



# Overview on Tool Abilities and Tool Performance
from: **Glanz, Oliver. “Bible Software on the Workbench of the Biblical Scholar: Assessment and Perspective.” Andrews University Seminary Studies 56, no. 1 (2018): 5–46.**


## Essential skills do not require digital tools
The ability to save space, to speed up reading, and to search for words improved the use of the classical tools (primary Hebrew and Greek texts, grammars, dictionaries) in research. The “tradition of doing” was boosted but not changed. While the process of textual analyses was sped up the nature of the process did not change.  This can be illustrated with the following example:
•	With the digitized Hebrew Bible an electronic concordance became available to the user. But the analog restriction of looking up words only is essentially not overcome by Bible Software. While it is possible to search not only for individual words, but also for specific word combinations (e.g. genitive construction [e.g. “house of David”], attributive construction [e.g. “the great heavens”]), there is no support for investigating specific language usage. In concrete spoken and written language expressions, words generally do not contain meaning in themselves. It is the orchestrated combination of words (language pragmatics and valence) that generates the meaning for each word.  The dictionary definition of the word “bright” is not very helpful for understanding the different meanings the word “bright” has in concrete linguistic expressions such as these:
-	“Thomas got a bright idea?”
-	“The sun is very bright today.”
-	“This girl looks so bright!”

In order to discover the specific meaning of the word “bright” one would need to find out whether the word was used in spoken language (first or second person account, direct speech tenses) or formal language (third person account, narrating tense), with what type of subject the word “bright” appeared in the expression, with what punctuation the expression ended, etc.. 
The two-dimensionality of printed concordances did not equip the user to successfully ask these necessary questions. Filtering the multidimensionality of linguistic expression was not supported by the analog tools. This situation did not change with the digitization of these analog concordances. The analog limitation, then, was brought into the digital space of Bible Software. 

This example shows how exegetical methodology as the backbone of any textual interpretation process was not generally affected by the digitization in Biblical studies.  Electronic tools did not change the way we do things methodologically. It only changed the speed with which we do the things we do.

This becomes obvious when we compare textbooks for exegesis and biblical Hebrew published over the last 20 years. Almost none of them refer to Bible Software as a necessary tool to be used or as a tool that has changed our understanding of biblical languages or the processes of exegesis.  Improvements in exegetical methodology and language understanding over the last 20 years were generally not caused by the digital revolution but by the great accumulation of primary and secondary sources in printed form and by the hermeneutic reflections influenced by postmodern language philosophy and linguistics. 

Therefore, the conclusion can be drawn that while Bible Software made our traditional practice of dealing with ancient Biblical texts easier, we do not necessarily have to incorporate it into the workflow of the biblical scholar. Consequently, Bible Software did only have a superficial rather than a deep impact on the outcomes of biblical research.  This is obviously the reason why Bible Software is usually not used in the classroom for learning grammar or exegesis. When Bible Software is found in the modern classroom it is mostly used for displaying Hebrew or Greek source texts, similar to the use of overhead projectors in the 20th century classroom. This is another example for how the analog experience is very much present in the digitized world of the biblical scholar.

When commercial Bible Software was initially developed it was designed to digitally mirror the analog tools of theologians. With this imitation the limitation of the “original” was transported as well: access to the graphical information of the sequence of words (primary text, analytic dictionaries, grammars, dictionaries). Since the different texts were linked (primary texts with the morphology of analytic dictionaries, which in turn were matched with dictionary entries) information was made available in very helpful ways. However, the nature of that information was static and an efficient critical engagement with digitized data was impossible. For example:

-	Some Hebrew dictionaries show two entries for the verb גלה.  I-גלה has the meaning “reveal” and II-גלה has the meaning “deport”. Other Hebrew dictionaries do not regard גלה as having two homographs but rather treat the meanings “reveal” and “deport” as belonging to the very same verbal root. As a user of Bible Software, one would like to investigate the biblical text in order to find out under what conditions what type of meaning is generated by that particular Hebrew verb. The user’s query will have to do more than just looking up all occurrences of the word. Questions like “Who is using the word?”, “Who is revealed and what is revealed?”, “Who is deported and who is deporting?” have to be asked. These questions, however, transcend the word boundary that came with the digitization of analog tools. Therefore, a critical investigation of dictionaries is as labor intensive as it was without Bible Software.

One could say that with Bible Software the user had access to lexical and grammatical interpretation but not to primary data allowing for digitally supported independent studies that involved word-boundary transcending linguistic levels (phrases, clauses, sentences, text-grammar).

As a consequence, one can see that with the growing sales of Bible Software user’s Bible interpretation was increasingly based upon lexical insights (and not on insights of particular language use - see my earlier “bright” example). Educators had to warn users of Bible Software that the “just-one-click-away-word-meaning” does not necessarily help to understand a particular Bible passage.

Electronic tools were not able to highlight information of a language pragmatic nature that goes beyond morphology. The rather subjective art of searching for keywords and rhetorical patterns was not balanced out with the search for language-pragmatic patterns and text-grammatical elements (i.e. the grammatical features that help “build” a text [e.g. relative pronouns like “which” that introduce relative clauses]), generating reproducible, inter-subjective information.

In summary we can say that Bible Software has improved (and changed) the techniques of publishing texts, the techniques of storing, and the techniques of distributing biblical texts. Both text producers (scribes) and text administrators (librarian) have greatly benefitted from modern electronic tools in their work.  However, in comparison, modern electronic tools in the form of Bible Software hardly supported the analytic eye of the scholar (method). The core business, namely the craft of analyzing ancient texts, was only insignificantly affected by electronic media. Referring to digital humanities in general Van Rompay put it this way:

>Admittedly, an impressive number of new texts have been published in the recent decade, but when it comes to the basic tools of language and literature, it is difficult to argue that the present-day student is much better off than her or his fellow students of eighty or hundred years ago.

#### Why does Abimelech's speech get introduced twice?



Isn't the text-syntactical construction in Gen 20:9-10 weird? Why should the speaker and addressee be introduced for a second time when the same speaker is still speaking to the same addressee?

````python
TwoSpeachesSameSpeaker1 = '''
chapter
        c1:clause domain=N
            phrase function=Pred
                word lex=DBR[|QR>[|>MR[
            phrase function=Subj
                speakerA:word sp=subs
            phrase function=Cmpl
                addresseeA:word
        c2:clause domain=Q
        c3:clause domain=N
            phrase function=Pred
                word lex=DBR[|QR>[|>MR[
            phrase function=Subj
                speakerB:word lex*
            phrase function=Cmpl
                addresseeB:word lex*

c1 <1: c2
c2 <50: c3
c1 < c3

speakerA .lex=lex. speakerB
addresseeA .lex=lex. addresseeB
'''
TwoSpeachesSameSpeaker1 = BHS.search(TwoSpeachesSameSpeaker1)
BHS.table(TwoSpeachesSameSpeaker1, start=1, end=50, condensed=True)
````

````python
# Dynamic programming implementation of LCS problem 
  
# Returns length of LCS for X[0..m-1], Y[0..n-1]  
def lcsreturn(X, Y, m, n): 
    L = [[0 for x in range(n+1)] for x in range(m+1)] 
  
    # Following steps build L[m+1][n+1] in bottom up fashion. Note 
    # that L[i][j] contains length of LCS of X[0..i-1] and Y[0..j-1]  
    for i in range(m+1): 
        for j in range(n+1): 
            if i == 0 or j == 0: 
                L[i][j] = 0
            elif X[i-1] == Y[j-1]: 
                L[i][j] = L[i-1][j-1] + 1
            else: 
                L[i][j] = max(L[i-1][j], L[i][j-1]) 
  
    # Following code is used to print LCS 
    index = L[m][n] 
  
    # Create a character array to store the lcs string 
    lcs = [""] * (index+1) 
    lcs[index] = "" 
  
    # Start from the right-most-bottom-most corner and 
    # one by one store characters in lcs[] 
    i = m 
    j = n 
    while i > 0 and j > 0: 
  
        # If current character in X[] and Y are same, then 
        # current character is part of LCS 
        if X[i-1] == Y[j-1]: 
            lcs[index-1] = X[i-1] 
            i-=1
            j-=1
            index-=1
  
        # If not same, then find the larger of two and 
        # go in the direction of larger value 
        elif L[i-1][j] > L[i][j-1]: 
            i-=1
        else: 
            j-=1

    # Return allows me to store it in the df as a column entry
    return(""  + "".join(lcs))
````

##### Adding statistical information
When having a large dataset we would like to filter out those LCS results that embody a high matching value between the first word (in our case the Object) and the second word (in our case the Predicate). In order to allow for such filtering we need additional information added in additional columns. Lets first add information about the string length of each `X` and `Y` word.

````python
# Adding character length information about the Object-word
PredObjcWordLists['Objclength']=PredObjcWordLists.apply(lambda row : len(row['TEXT4']), axis = 1)
# Adding character lenght information about the Predicate-word
PredObjcWordLists['Predlength']=PredObjcWordLists.apply(lambda row : len(row['TEXT6']), axis = 1)
# Adding character lenght information about the LCS-result
PredObjcWordLists['LCSlength']=PredObjcWordLists.apply(lambda row : len(row['lcsresults']), axis = 1)
PredObjcWordLists.head(4)
````

For our research we are interested in displaying:
1. the matching ratio between object-word-lenght and LCS-length
```python
#applied to row#2

=40%
```

2. the matching ratio between predicate-word-lenght and the LCS-length
```python
#applied to row#2
11/100*4
=44%
```

3. the mean ratio of object+predicate-word-lenght andthe LCS-lenght
```python
#applied to row#2
(10+11)/2/100*4
=42%
```


````python
 def percentage(v1, v2):
      pctg = float(v1/100 * v2)
      return(pctg)
````

````python
PredObjcWordLists['LCSmatchObjc']=PredObjcWordLists.apply(lambda row : percentage(row['Objclength'], row['LCSlength']), axis = 1)
PredObjcWordLists.head(6)
````

````python
PredObjcWordLists['LCSmatchPred']=PredObjcWordLists.apply(lambda row : percentage(row['Predlength'], row['LCSlength']), axis = 1)
PredObjcWordLists.head(6)
````

````python
 def meanpercentage(v1, v2, v3):
      mean = float((v1+v2)/2)
      meanpctg = float(100/mean * v3)
      return(meanpctg)
````

````python
PredObjcWordLists['LCSObjcPredMeanMatch']=PredObjcWordLists.apply(lambda row : meanpercentage(row['Objclength'], row['Predlength'], row['LCSlength']), axis = 1)
PredObjcWordLists.head(6)
````

````python
PredObjcWordLists.sort_values(['LCSObjcPredMeanMatch'], ascending=[False])
````

````python
PredObjcWordLists.where(PredObjcWordLists['LCSObjcPredMeanMatch'] > 0.4)
````
