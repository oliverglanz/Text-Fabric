### לְפִ֥י דְבָרִֽי
What is probably more interesting is the prepositional phrase that follows the double conjunction "for the mouth of my word". Lets see where else we find phrases that contain mouth and word:

````python
MouthWord='''
phrase
    word lex=PH/
    word lex=DBR/
'''
MouthWord=OT.search(MouthWord)
OT.table(MouthWord, colorMap={2:'cyan', 3:'magenta'})
````

A quick look at the data shows that with the exception of [1 Kgs 17:1](https://www.stepbible.org/?q=version=ESV|version=OHB|reference=1Kgs.17.1&options=UVGVNH&display=COLUMN) the construction appears only in the Pentateuch. It is only there where the phrase consists of the order:

```
phrase
 w1=word lex=PH/
 w2=word lex=DBR/
w1 < w2
```

In all the other cases DBR/ precedes PH/:

```
phrase
 w1=word lex=DBR/
 w2=word lex=PH/
w1 < w2
```

Thus, although there is no strong distribution of the phenomenon, one could argue that the formulation in 1 Kgs 17:1 is typically Mosaic.* It is Mosaic as it shares the following features:
1. PH/ is in construct relation to DBR/. This is not the case outside of the Pentateuch.
2. The Pentateuch knows ONLY of the order PH/ => DBR/ as it does nowhere entertain the order DBR/ => PH/.
3. DBR/ establishes alway the end of the construction chain.
4. DBR/ is always determined (article attached).


However, 1 Kgs 17:1 differes in two aspects from the cases in the Pentateuch:
1. the PH/ in 1 Kgs 17:1 receives its determination through the 1sgC suffix attached to DBR/
2. PH/ is goverend by the preposition L!

In order to explore the phenonenon further it makes sense to look at other cases where:
1. Explore the function of L-PH/ constructions.
2. PH/ stands in construct with another noun
3. Where DBR/ is the end of a construct chain.

We explore these constructions below.

*Our analysis must exclude the case in [1 Sam 15:24](https://www.stepbible.org/?q=version=ESV|version=OHB|reference=1Sam.15.24&options=UVGVNH&display=COLUMN) as it does not show PH/ and DBR/ in a construct relationship and shows also a different phrase architecture (extended object-phrase).

#### PH/ goverend by the L preposition

````python
L_PH='''
phrase
 w1:word lex=L
 w2:word lex=PH/

w1 <: w2
'''

L_PH=OT.search(L_PH)
OT.table(L_PH, colorMap={2:'cyan',3:'magenta'})
````

One can easily observe that L-PH appears usually in idiomatic expression. The strongest distribution has the construction L-PH/-XRB/ rendered "by the opening of sword" (ESV: "with the edge of the sword"). Often it carries the meaning of "meassure" or "amount"  ("the amount that goes through your mouth") as in [Ex 12:4](https://www.stepbible.org/?q=version=ESV|version=OHB|reference=1Sam.15.24&options=UVGVNH&display=COLUMN) "And if the household is too small for a lamb, then he and his nearest neighbor shall take according to the number of persons; **according to what each can eat** (אִישׁ לְפִי אָכְל) you shall make your count for the lamb." (ESV)

The meaning of "meassure"/"amount"/"proportion" is also present in constructions with CNH/ as in [Lev 25:16](https://www.stepbible.org/?q=version=ESV|version=OHB|reference=Lev.25.16|version=KJVA&options=UVGVNH&display=COLUMN): "**According to the multitude of years** (לְפִי ׀ רֹב הַשָּׁנִי) thou shalt increase the price thereof, and **according to the fewness of years** (וּלְפִי מְעֹט הַשָּׁנִ) thou shalt diminish the price of it: for according to the number of the years of the fruits doth he sell unto thee." (KJV)

See also [Lev 27:16](https://www.stepbible.org/?q=version=ESV|version=OHB|version=KJVA|reference=Lev.27.16&options=UVGVNH&display=COLUMN): "If a man dedicates to the  Lord part of the land that is his possession, then the valuation shall be in proportion to its seed (לְפִ֣י זַרְעֹ֑ו). A homer of barley seed shall be valued at fifty shekels of silver." (ESV)

As the formualtion in 1 Kgs 17:1 is exceptional one could explore the following options:

1. As the idiomatic usages shows there is a personification of the absolute element in the construct chain (swords have mouths, years have mouths). Kings could play with this idom and personifiy the word as having a mouth.
2. One finds also the formulations "by his word" (e.g. Ex 28:32). 1 Kings could extend this formulations by putting the stress on the actual content of the words rather than the organ of the mouth: the mouth of my words. This would be a reversal of the normal construction "words of my mouth" (e.g. Jer 9:19, Psa 36:4, Prov 18:4, Eccl 10:12, 10:13)

#### PH/ at the beginning of construct chains

````python
PHstart='''
phrase
 w1:word lex=PH/ st=c
 w2:word lex sp=subs
w1 < w2
'''

PHstart=OT.search(PHstart)
OT.table(PHstart, colorMap={2:'cyan',3:'magenta'})
````

#### DBR/ at the end of a construct chain

````python
DBRend='''
phrase
 w1:word sp=subs st=c
 w2:word lex=DBR/
w1 < w2
'''

DBRend=OT.search(DBRend)
OT.table(DBRend, colorMap={2:'cyan',3:'magenta'})
````
